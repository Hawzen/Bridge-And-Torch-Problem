```python
from bridge_problem import *
import pandas as pd
```

```python
test_cases = read_test_cases("test_cases_hard.txt")
# prob = BridgeProblem(case)
df, sol = generate_report(test_cases, df_only=False)
```

```python
df
```

```python
scatters = {dimension: pd.DataFrame(columns=df.columns) for dimension in ["Solution Cost", "Search Cost", "Space Requirement", "Time (s)"]}
for (test_case, metric), algo in df.iterrows():
    scatters[metric].loc[test_case] = algo

from itertools import count
for val in scatters.values():
    cnt = count()
    val.index = val.index.map(lambda x: f"T{next(cnt)+1}")
```

```python
import plotly.express as px 
import plotly.graph_objects as go
from plotly.subplots import make_subplots

size = 10
colors = px.colors.qualitative.Pastel[:3]
fig = make_subplots(rows=2, cols=2)
for i, dim in enumerate(scatters.keys()):
    r, c = i%2+1, 1 if i < 2 else 2
    fig.add_traces([
        go.Scatter(x=scatters[dim].index, y=scatters[dim]["UCS"], name=f"UCS", legendgroup=f"UCS", mode="lines+markers", marker=dict(size=size, color=colors[0]), showlegend=i==0),
        go.Scatter(x=scatters[dim].index, y=scatters[dim]["IDS"], name=f"IDS", legendgroup=f"IDS", mode="lines+markers", marker=dict(size=size, color=colors[1]), showlegend=i==0),
        go.Scatter(x=scatters[dim].index, y=scatters[dim]["A*"], name=f"A*", legendgroup=f"A*", mode="lines+markers", marker=dict(size=size, color=colors[2]), showlegend=i==0), 
    ], rows=r, cols=c)
    
    fig.update_xaxes(
        title_text="Test Case",
        row=r,
        col=c
    )
    
    fig.update_yaxes(
        title_text=dim,
        row=r,
        col=c
    )
    
fig.update_layout(
    xaxis_title="Test Case",
#     font= dict(
#         family="Courier New",
#         size=20,
        
#     ),
#     title= "Algorithm Comparison"
#     legend_title="Algorithm"
)
fig.write_html("file.html")
```
